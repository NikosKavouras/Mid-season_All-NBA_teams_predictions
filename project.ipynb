```python
#import all necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math

%matplotlib inline

import matplotlib as mpl
mpl.rc("savefig", dpi=100) # Adjust for higher-resolution figures
```

```python
#read all three files
stats =pd.read_csv('2021stats.csv',header=0)
adv_stats =pd.read_csv('2021advanced.csv',header=0)
all_NBA =pd.read_csv('2021all-NBA.csv',header=0)
```

```python
#merge the three files
tempdf=stats.merge(adv_stats, left_on='Player-additional', right_on='Player-additional')
finaldf=tempdf.merge(all_NBA, left_on='Player-additional', right_on='Player-additional')
```

```python
#print all column names
lista=[]
for col in finaldf.columns:
    lista.append(col)
print(lista)
```

```python
#drop duplicate columns and display All-NBA selections
uniquedf=finaldf.drop(columns=['Rk_x','Rk_y', 'Pos_y', 'Age_y', 'Tm_y', 'G_y', 'MP_y','Rk'])
display(uniquedf[uniquedf['All-NBA']>0])
```

```python
#We are only going to check the importance of variables that make sense for the selection process
#That is, we are excluding stats that are very specific or of minor importance for assessing the performance of the players
#Finally, defensive statistics are also excluded, because they are used in the selection process of the All-Defensive Teams of the NBA.
impdf=uniquedf[['G_x', 'MP_x','TRB', 'AST', 'PTS', 'PER', 'USG%', 'OWS', 'WS', 'OBPM', 'BPM', 'VORP']]
labeldf=uniquedf["All-NBA"]
y = labeldf.to_numpy().reshape(605,1)
```

```python
#import libraries needed to run regression
from sklearn import linear_model
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

#scale data
scale = StandardScaler()
scaledX = scale.fit_transform(impdf)

#Run logistic regression and print coefficients
regr = linear_model.LogisticRegression()
coef = regr.fit(scaledX, y).coef_

for r in range(len(coef.flatten())):
    print(impdf.columns.values.tolist()[r])
    print(coef.flatten().tolist()[r])
    
#From this analysis, weconclude that the most significant predictors are
#PTS(=Points scored), WS(=Win shares), and VORP(=Value over replacement player)
```

```python
#Now we run the logistic regression again, for the best predictors
newdf=impdf[['WS', 'VORP', 'PTS']]
scaledX2 = scale.fit_transform(newdf)
regr = linear_model.LogisticRegression()
coef2 = regr.fit(scaledX2, y).coef_

for r in range(len(coef2.flatten())):
    print(newdf.columns.values.tolist()[r])
    print(coef2.flatten().tolist()[r])
```

```python
def make_scatter_plot(df, x, y, hue="All-NBA",
                      palette={0: "blue", 1: "red"},
                      size=5):
    sns.lmplot(x=x, y=y, hue=hue, data=df, palette=palette,
               fit_reg=False)

mpl.rc("savefig", dpi=120) # Adjust for higher-resolution figures

make_scatter_plot(finaldf, 'VORP','WS')
```

```python
#We create and print the confusion matrix
predictions=regr.predict(scaledX2)
cm = metrics.confusion_matrix(y, predictions)
display(pd.DataFrame(cm, columns = ['Prediction: Non All-NBA','Prediction: All-NBA']))
```

```python
#We print the acuraccy of our predictions
print('Acuraccy: ',round(regr.score(scaledX2,y)*100,4),"%")
```
